```python
from sklearn.model_selection import train_test_split
from matplotlib.pyplot import matplotlib
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
scaler = StandardScaler() # standardized feature by removing mean and scaled to unit variance 
scaler.fit(data)

X_scaled = scaler.transform(data)


print (X_scaled.min(axis=0))


```

```python
n_components=3
pca = PCA(n_components=n_components) # instantiate the PCA and keep the first n components
pca.fit(X_scaled)
# now transform 

x_pca=pca.transform(X_scaled)

# check the shape 

print (x_pca.shape)
#print pca.explained_variance

ex_variance=np.var(x_pca,axis=0)

ex_variance_rat=ex_variance/np.sum(ex_variance)

print (ex_variance_rat)

```

```python
#This is how you generate samples:
#First, sample in the PCA latent space, i.e., coordinates along the principal axis vectors
nSamples=20
latentMean=np.zeros(n_components)
latentStd=np.sqrt(ex_variance)
latent=np.random.normal(latentMean,latentStd,[nSamples,n_components])
#print(latent)
#Now, we can convert back to the original space
sampled=pca.inverse_transform(latent)
print(sampled)
```

```python
new=scaler.inverse_transform(sampled)	
print(new)
```
